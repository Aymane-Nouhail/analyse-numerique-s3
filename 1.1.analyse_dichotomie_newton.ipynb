```python
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return x**3-x-3
def g(x):
    return 3*x**2-1
```

```python
def dichotomie(f,a,b,n): #f est croissante
    alpha = a
    for i in range(n+1):
        alpha = (a+b)/2
        if f(alpha)>0:
            b = alpha
        else:
            a = alpha
    return alpha
print(dichotomie(f,1,2,1000))
```

```python
def newton(f,g,a,n): #g = f'
    for i in range(n):
        a = a - f(a)/g(a)
    return a
print(newton(f,g,2,1000))
```

```python
def dichotomie2(f,a,b,eps):
    m = int(math.log((b-a)/eps,2))
    return dichotomie(f,a,b,m)
print(dichotomie2(f,1,2,0.0000001))
```

```python
n=18
#y = [dichotomie(f,-5,5,i) for i in range(100)]
y1 = [newton(f,g,1.5,i) for i in range(n)]
x = [i for i in range(n)]
y2 = [dichotomie(f,1,2,i) for i in range(n)]
plt.plot(x,y1,'ro',c='r',label = "Newton") 
plt.plot(x,y2,'ro',c='b',label = "Dichototmie")
plt.xticks([i for i in range(n)])
plt.grid()
plt.legend()
plt.show()
```
