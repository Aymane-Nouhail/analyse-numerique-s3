```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dichotomie(f,a,b,n): #f est croissante
    alpha = a
    for i in range(n+1):
        alpha = (a+b)/2
        if f(alpha)>0:
            b = alpha
        else:
            a = alpha
    return alpha

def newton(f,g,a,n): #g = f'
    for i in range(n):
        a = a - f(a)/g(a)
    return a
```

```python
def y(t, to, a, v0, g):
    return (v0*math.sin(a)/to + g/to**2)*(1-math.exp(-to*t))-t*g/to

def yfinal(t):
    return -y(t,0.01,math.pi/3,100,9.8)
#################################################
def x(t, to, a, v0, g):
    return v0*math.cos(a)/to * (1-math.exp(-to*t))

def xfinal(t):
    return x(t,0.01,math.pi/3,100,9.8)
```

```python
T = dichotomie(yfinal,12,18,1000) #T est la racine non-nul de y.
print("T = ",T)
print("y(T) = ",yfinal(T))
L = xfinal(T) #L est l'image de T par x.
print("L = ",L)
tm= math.log(0.01*100*math.sin(math.pi/3)/(9.8)+1,math.e)/(0.01)
ym = y(tm,0.01,math.pi/3,100,9.8)
plt.plot(tm,ym,marker='*' ,color='g')
plt.plot(T,0,marker='*',color='r')
print(tm,ym)
x1=np.linspace(0,18,100)
y1 = [y(t,0.01,math.pi/3,100,9.8) for t in x1]
y2 = [x(t,0.01,math.pi/3,100,9.8) for t in x1]
plt.plot(x1,y1)
#plt.plot(x1,y2)
plt.annotate('altitude maximale',(tm-3,ym-25),weight = "bold",c = 'g')
plt.annotate('point de chute',(T-6.3,-14),weight = "bold",c = 'r')
plt.grid()
plt.show()
```
